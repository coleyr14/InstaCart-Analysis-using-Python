### Line Chart

```python
np.random.seed(4)
```

```python
# create list holding T/F values to the test np.random.rant()<=0.7
dev = np.random.rand(len(customers_orders))<=0.7
```

```python
# store 70% of the sample in dataframe big
big = customers_orders[dev]
```

```python
# store 30% of the sample in datafram small
small = customers_orders[~dev]
```

```python
len(customers_orders)
```

```python
len(big) + len(small)
```

```python
# reduce size by only using columns needed
df = small[['order_dow','prices']]
```

```python
#using 30%
line_order_dow_prices = sns.lineplot(data = df, x = 'order_dow',y = 'prices')
```

# Step 3

```python
#create histogram for what hours during the day orders are placed
hist_order_hod = customers_orders['order_hour_of_day'].plot.hist(bins = 25)
```

as shown in the above histogram - the x axis shows the hour (time) during the day (0=12am, 17=5pm, etc), the y axis shows the frequency of orders placed at those hours(times) - the most frequent hours during the day (24 hour period) begin around 9am and continue through to 4pm minus the pause that happens around 12pm noon

# Step 6

```python
# will use data sample from previous step - 30%
# reduce size by only using columns needed
df_3 = small[['Age','n_dependants']]
```

```python
line_age_depend = sns.lineplot(data = df_3, x = 'Age',y = 'n_dependants')
```

the above line chart shows the age starting at 18 up to 81 (this numbers were given from the min and max in part 1) and along the y axis is the number of dependants - this line chart really isn't showing any correlation for age and number of dependants in a family, also having a non integer for dependants does not help find a correlation since dependants really should be a whole integer since it is representing a person in a family

# Step 7
